```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
# pred_probas = np.random.rand(100000)
# y_true = np.random.randint(0,2,100000)
X,y = make_classification(n_samples=200000, n_features=5, n_informative=2)
```

```python
#  df = pd.DataFrame({'pred_proba': pred_probas, 'target': y_true})
#  df=df.rename(columns={'pred':'pred_proba', 'label':'target' })
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'feature3':X[:,2],'feature4':X[:,3],'feature5':X[:,4],'target':y})
# df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.5,random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
y_pred_dt = model_dt.predict(X_test)
```

```python
model_rfc = DecisionTreeClassifier()
model_rfc.fit(X_train,y_train)
y_pred_rfc = model_rfc.predict(X_test)
```

```python
model_cbr = CatBoostClassifier()
model_cbr.fit(X_train, y_train)
y_preds_cbr = model_cbr.predict(X_test)
```

## Time Complexity

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
import time

start_time = time.time()
print("Accuracy",accuracy_score(y_test,y_pred_dt))
end_time = time.time()
print("Time Taken",end_time - start_time)
```
